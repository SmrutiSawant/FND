```python
import re
import string

def wordopt(text):
    """
    Preprocesses a given text by removing unnecessary elements such as 
    punctuation, URLs, numbers, and special characters.
    Args:
    text (str): The input text string to be cleaned.
    Returns:
    str: The cleaned and processed text.
    """
    # Convert text to lowercase to ensure uniformity
    text = text.lower()
    # Remove text within square brackets (e.g., citations like [1], [example])
    text = re.sub(r'\[.*?\]', '', text)  
    # Remove non-word characters (anything that is not a letter or number)
    text = re.sub(r"\W", " ", text)    
    # Remove URLs (http, https, or www)
    text = re.sub(r'https?://\S+|www\.\S+', '', text)  
    # Remove HTML tags (e.g., <p>, <br>)
    text = re.sub(r'<.*?>+', '', text)  # (b'' is incorrect, removed that)
    # Remove punctuation marks
    text = re.sub(r'[%s]' % re.escape(string.punctuation), '', text)
    # Remove words containing numbers (e.g., "word123" or "123word456")
    text = re.sub(r'\w*\d\w*', '', text)
    
    return text
```

```python
import re
import string

def wordopt(text):
    """
    Preprocesses a given text by removing unnecessary elements such as 
    punctuation, URLs, numbers, and special characters.
    Args:
    text (str): The input text string to be cleaned
    Returns:
    str: The cleaned and processed text.
    """
    # Convert text to lowercase to ensure uniformity
    text = text.lower()
    # Remove text within square brackets (e.g., citations like [1], [example])
    text = re.sub(r'\[.*?\]', '', text)
    # Remove non-word characters (anything that is not a letter or number)
    text = re.sub(r"\W", " ", text) 
    # Remove URLs (http, https, or www)
    text = re.sub(r'https?://\S+|www\.\S+', '', text)
    # Remove HTML tags (e.g., <p>, <br>)
    text = re.sub(r'<.*?>+', '', text)  # (b'' is incorrect, removed that)
    # Remove punctuation marks
    text = re.sub(r'[%s]' % re.escape(string.punctuation), '', text)
    # Remove words containing numbers (e.g., "word123" or "123word456")
    text = re.sub(r'\w*\d\w*', '', text)
    return text
```

```python
# Separating features (text) and target labels (class)
x = data['text']  # Independent variable (news content)
y = data['class']  # Dependent variable (0 = Fake news, 1 = True news)
```

```python
x_train, x_test, y_train, y_test = train_test_split(x,y,test_size = 0.25)
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer
# Initializing the TF-IDF Vectorizer
vectorization = TfidfVectorizer()
# Fitting the vectorizer to the training data and transforming it into numerical format
xv_train = vectorization.fit_transform(x_train)
# Transforming the test data using the same fitted vectorizer
xv_test = vectorization.transform(x_test)
```

```python
from sklearn.linear_model import LogisticRegression# Initializing the Logistic Regression model
LR = LogisticRegression()

# Training the model using the TF-IDF transformed training data
LR.fit(xv_train, y_train)
```

```python
# Function to return the label based on prediction output
def output_label(n): 
    if n == 0:
        return "Fake News"
    elif n == 1:
        return "Not A Fake News"
    
# Function for manual testing of a given news article
def manual_testing(news):
    # Create a DataFrame for the input news article
    testing_news = {"text": [news]}
    new_def_test = pd.DataFrame(testing_news)
    
    # Apply text preprocessing (wordopt function should be defined)
    new_def_test['text'] = new_def_test["text"].apply(wordopt)
    
    # Transform the text using the pre-trained vectorizer
    new_x_test = new_def_test["text"]
    new_xv_test = vectorization.transform(new_x_test)
    
    # Make predictions using Logistic Regression and Decision Tree models
    pred_LR = LR.predict(new_xv_test)  # Logistic Regression Prediction
    pred_DT = DT.predict(new_xv_test)  # Decision Tree Prediction
    
    # Print predictions from both models
    return print("\n\nLR Prediction: {} \nDT Prediction: {}".format(
        output_label(pred_LR[0]),  # Convert prediction output to label
        output_label(pred_DT[0])   # Convert prediction output to label
    ))
```
